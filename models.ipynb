```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import kagglehub

path = kagglehub.dataset_download("manishkc06/mobile-price-prediction")

print("Path to dataset files:", path)
```

```python
import os

file = os.listdir(path)
new_path = os.path.join(path, file[0])
data = pd.read_csv(new_path)
```

```python
os_mappings = {
    "Android Pie 9.0": "Android Pie 9",
    "Android Oreo 8.0": "Android Oreo 8",
    "Android Nougat 7.1.2":"Android Nougat 7.1",
    "Android Marshmallow 6.0.1": "Android Marshmallow 6",
    "Android Lollipop 5.4.1":"Android Lollipop 5.1",
    "Android Lollipop 5.1.1": "Android Lollipop 5.1",
    "Android Nougat 7.1.1": "Android Nougat 7",
    "Android Oreo 8.1.0":"Android Oreo 8",
    "Android Oreo 8.1": "Android Oreo 8",
    "Android Nougat 7.1":"Android Nougat 7"
}
```

```python
data['os']=data['os'].replace(os_mappings)
```

```python
data['mobile_color'].value_counts()
```

```python
data['mobile_color'] = data['mobile_color'].apply(
    lambda x: 'Black' if 'black' in x.lower() else 
              ('Green' if 'green' in x.lower() else 
               ('Red' if 'red' in x.lower() else 
                ('Gold' if 'gold' in x.lower() else 
                 ('Blue' if 'blue' in x.lower() else 
                  ('Grey' if 'grey' in x.lower() else 
                   ('Sapphire' if 'sapphire' in x.lower() else 
                    ('Silver' if 'silver' in x.lower() else 
                     ('Mint' if 'mint' in x.lower() else 
                      ('Purple' if 'purple' in x.lower() else 
                       ('White' if 'white' in x.lower() else "Other")))))))))))


data.drop(['mobile_color'], axis=1, inplace=True)
```

```python
data['dual_sim'].value_counts()
```

```python
data.drop(['dual_sim'], axis = 1, inplace=True)
```

```python
data['network'].value_counts()
```

```python
data.drop(['bluetooth'], axis=1, inplace=True)
```

```python
data['mp_speed']=data['mp_speed'].apply(lambda x: float(x[:-4]))
```

```python
data['int_memory']=data['int_memory'].apply(lambda x: float(x[:-3]))
```

```python
data['mob_weight'] = data['mob_weight'].apply(lambda x: float(x[:-2]))
```

```python
data['mob_depth'] = data['mob_depth'].apply(lambda x: float(x[:-3]))
data['mob_height'] = data['mob_height'].apply(lambda x: float(x[:-3]))
data['mob_width'] = data['mob_width'].apply(lambda x: float(x[:-3]))
```

```python
data['battery_power'] = data['battery_power'].apply(lambda x: float(x[:-4]))
data['f_cam'] = data['f_cam'].apply(lambda x: float(x.split()[0][:-2]))
```

```python
data['ram'] = data['ram'].apply(lambda x: float(x[:-3]))
```

```python
import re

# Clean the resolution string by removing non-numeric characters and split by ' x'
def extract_resolution(res):
    res = re.sub(r'[^0-9 ]', '', res)  
    parts = res.split('  ')
    if len(parts) == 2:
        return int(parts[0]), int(parts[1])
    else:
        return None, None

# Apply the function to the 'resolution' column
data[['reso_height', 'reso_width']] = data['resolution'].apply(lambda x: pd.Series(extract_resolution(x)))

```

```python
data.drop(['resolution'], axis=1,inplace=True)
```

```python
data['disp_size'] = data['disp_size'].apply(lambda x: float(x[:5]))
```

```python
data['mobile_price'] = data['mobile_price'].apply(lambda x: float(x[1:].replace(',', '.')))
```

```python
data['mobile_name'] = data['mobile_name'].apply(lambda x: x.split()[0])
```

```python
data['p_cam'] = data['p_cam'].apply(lambda x: (x[:3]))
```

```python
def extract_cam(cam_p):
    res = re.sub(r'[^0-9]','', cam_p)
    return int(res)

data['p_cam'] = data['p_cam'].apply(lambda x: extract_cam(x))
```

```python
data.drop(['network'], axis = 1, inplace=True)
```

```python
data
```

```python
data['mobile_name'].value_counts()
```

```python
name_new_columns = pd.get_dummies(data['mobile_name'], prefix='name_')
```

```python
name_new_columns
```

```python
os_new_columns = pd.get_dummies(data['os'], prefix='os_')
```

```python
num_cores_new_columns = pd.get_dummies(data['num_cores'], prefix='num_cores_')
```

```python
data = pd.concat([data, os_new_columns, name_new_columns, num_cores_new_columns], axis=1)
data.drop(['os','num_cores','mobile_name'], axis=1, inplace=True)
```

```python
data['mobile_price'] = data['mobile_price'].apply(lambda x: float(''.join(str(x).split('.'))))
```

```python
data = data.dropna()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split


model = LinearRegression()

X,y = data.drop(['mobile_price'], axis=1), data['mobile_price']
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import root_mean_squared_error, mean_absolute_error
```

```python

y_pred = model.predict(x_test)
results = root_mean_squared_error(y_test,y_pred)
```

```python
results
```

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor()
rfr.fit(x_train, y_train)
y_pred = rfr.predict(x_test)
results = root_mean_squared_error(y_test,y_pred)
```
